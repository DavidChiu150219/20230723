```python
import sympy
import matplotlib.pyplot as plt

t,s = sympy.symbols('t,s')
a = sympy.symbols('a',real=True,positive=True)

f=sympy.exp(-a*t)
g=sympy.exp(-a*t)*t**2


def L(f):
    return sympy.laplace_transform(f,t,s, noconds=True)

def invL(F):
    return sympy.inverse_laplace_transform(F,s,t)

F=L(f)
G=L(g)
display(F)
display(invL(F))

display(G)
display(invL(G))
```

```python
from scipy.integrate import quad_vec
import numpy as np
import matplotlib.pyplot as plt
alpha = np.linspace(0.0, 2.0, num=30)
f = lambda x: x**alpha
x0, x1 = 0, 2
y, err = quad_vec(f, x0, x1)
plt.plot(alpha, y)
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$\int_{0}^{2} x^\alpha dx$")
plt.show()
```

```python
from sympy import*
from sympy.plotting import plot
x,n, p=var('x,n, p')
def besselj(p,x):
    return summation(((-1)**n*x**(2*n+p))/(factorial(n)*gamma(n+p+1)*2**(2*n+p)),[n,0,oo])
st="J_{p}(x)"
p1=plot(besselj(0,x),(x,-20,20),line_color='b',title=' $'+st+ '$',show=False)    
p2=plot(besselj(1,x),(x,-20,20),line_color='g',show=False)  
p3=plot(besselj(2,x),(x,-20,20),line_color='r',show=False)
p4=plot(besselj(3,x),(x,-20,20),line_color='c',show=False)
p1.extend(p2)
p1.extend(p3)
p1.extend(p4)
p1.show()
```
